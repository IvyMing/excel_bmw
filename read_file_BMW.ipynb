```python
filename = r'C:\Users\明\Desktop\BMW\2022-04-04_09-12_beat_softwareverbund_I_K22_22-07_505_22w13_DE61.xlsx'
sheetname = 'Funkt_und_Bewertung_n_Standort'
```

```python
# DICTIONARY TO CLASSIFY GROUP BBASED ON KEYWORDS
# keyword : GROUP

KEYWORDS = {
    "ADAS" : "ADAS",
    "audio" : "AUDIO",
    "AV" : "AUDIO",
    "call" : "call"
    
}

def group(x):
    for g in KEYWORDS.keys():
        if g in x:
            return KEYWORDS[g]
    return "Default"
```

```python
def remove_time(x):
    y = x.split("\n")
    if ":" in y[0]:
        y[0] = ":".join(y[0].split(':')[1:])
    elif " " in y[0]:
        y[0] = " ".join(y[0].split(' ')[1:])
    
    return "\n".join(y)
    
```

```python
data = pd.read_excel(filename,sheet_name=sheetname) 
# X = pd.DataFrame(data, columns= ['Standort Name'])
# print (df)



df = pd.DataFrame(data, columns= ['Name EN Reportelement','AIDA IP RG','Standort RG','Standort Bemerkung','Standort Name'])
df = df[df['Standort Name'] == 'JAP']
groups = [group(x) for x in df['Name EN Reportelement']]
df['Standort Bemerkung'] = [remove_time(x) for x in df['Standort Bemerkung']]
cols = df.columns.tolist()
cols = ["Group"] + cols
df["Group"] = groups
df = df[cols]
print (df)
patth = r'C:\Users\明\Desktop\BMW\list0.xlsx'
df.to_excel(patth)
#to_pdf(df, filename="list7")
```
